```python
#import packages
import os
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

ini_data = pd.ExcelFile('DATA_VKR_FIN.xlsx')
findf = ini_data.parse('Fin_Data')
findata = findf[['Eq_TL', 'ShtDebt_Eq', 'Eq_TA', 'D_EBIT', 'Cash_TA', 'Cur_rat',
                'Tan_TA', 'ROE', 'ROA', 'ROCE', 'EBIT_IntEx', 'S_TA', 'CFO_S', 'CFO_OperPr',
                'lnTA', 'GPM', 'Pretax_margin']]
findata.head()
```

```python
desc_stat_fin = findata.describe()
desc_stat_fin.to_excel('desc_stat_fin.xlsx')
```

```python
qualdf = ini_data.parse('Qualitative_data (с нулями)')
qualdata = qualdf[['Comp_age', 'Corp_group', 'Ind_innov', 'M_founder', 'Man_change',
                  'Is_Europe', 'Is_Asia', 'Is_service', 'Is_SME', 'Is_LLC']]
desc_qualdata = qualdata.describe()
desc_qualdata.to_excel('desc_qualdata.xlsx')
```

```python
macro = ini_data.parse('Macrodata')
macrodata = macro[['Unempl_rat', 'Inf_FedReg']]
desc_macrodata = macrodata.describe()
desc_macrodata.to_excel('desc_macrodata.xlsx')
```
